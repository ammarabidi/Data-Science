```python
import seaborn as sns
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title("Line Plot of Tips vs Total Bill")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
sns.boxplot(x="total_bill", y="tip", data=tips)
plt.title("Line Plot of Tips vs Total Bill")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
```

```python
flights = sns.load_dataset("flights")
```

```python
flights
```

```python
pv = flights.pivot(index="month",columns= "year",values= "passengers")
```

```python
pv
```

```python
sns.heatmap(pv, annot=True, fmt="d", cmap="YlGnBu")    #fmt="d" formats the annotations as integers
plt.title("Heatmap of Passengers by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.tight_layout()
```
