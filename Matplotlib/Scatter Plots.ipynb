```python
import matplotlib.pyplot as plt
```

```python
# Sample data
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 90, 55, 60, 25, 75, 85, 90]
```

```python
#colors = ['red' if scores<50 else 'green' for scores in exam_scores]
sizes = [score *2 for score in exam_scores]  
plt.scatter(study_hours, exam_scores,c=exam_scores, cmap='viridis')
plt.colorbar(label='Exam Scores')
# Add labels
for i in range(len(study_hours)):
    plt.annotate(f'S{i+1}', (study_hours[i], exam_scores[i]-2.5)) #annotations for each point
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.grid(True)
plt.show()
```

```python
# Assume two groups: Class A and Class B
class_a_hours = [2, 4, 6, 8]
class_a_scores = [95, 15, 65, 85]
 
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]
```

```python
plt.scatter(class_a_hours, class_a_scores, color='blue', label='A')
plt.scatter(class_b_hours, class_b_scores, color='red', label='B')
plt.grid(True)
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')   
plt.title('Class A vs Class B Exam Scores')
plt.legend(title='Class', loc='upper left')
```
